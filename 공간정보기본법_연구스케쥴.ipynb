## 연구 개요

국외 플랫폼 생태계 현황 분석 및 온쇼어 보안체계 연구 (박진우 교수 연계)

### 조사 대상 플랫폼
| 플랫폼 | 운영사 | 비고 |
|--------|--------|------|
| Google Maps | Google (Alphabet) | 주요 대상 |
| Apple Maps | Apple | |
| A-Map (高德地图, 고덕지도) | Alibaba | 조사 대상 포함 확정 |
| Yandex Maps | Yandex | |

### 조사 항목
1. 지도와 연계 서비스 현황 (호텔예약, 식당예약, 우버 등)
2. API 서비스 현황
3. 보안 체계 관련 사례 분석
