## 연구 개요

국외 플랫폼 생태계 현황 분석 및 온쇼어 보안체계 연구 (박진우 교수 연계)

### 조사 대상 플랫폼
| 플랫폼 | 운영사 | 비고 |
|--------|--------|------|

| A-Map (高德地图, 고덕지도) | Alibaba | 조사 대상 포함 확정 |
| Yandex Maps | Yandex | |

### 조사 항목
1. 지도와 연계 서비스 현황 (숙박, 식당 예약, 차량공유/택시, 항공편 등)
2. API 서비스 현황
3. 온쇼어 이슈와의 연관성 분석

## 기존 조사 통합 결과 요약 (251126 문서 기반)

### Google Maps 서비스 카테고리
차량 / 대중교통 / 자전거 / 도보 / 차량공유·택시 / 항공편(Google Flights)

### 국가별 서비스 품질 비교
| 국가 | 차량 | 대중교통 | 자전거 | 도보 | 차량공유/택시 | 항공편(웹) |

| 미국 | 높음 | 높음 | 높음 | 높음 | 높음 | 높음 |
| 대한민국 | **낮음** | 중간 | **낮음** | **낮음** | **낮음** | 중간 |
| 스웨덴 | 높음 | 높음 | 높음 | 높음 | 낮음 | - |
| 영국 | 높음 | 높음 | 높음 | 높음 | 높음 | - |
| 인도 | 높음 | 높음 | 낮음 | 높음 | 낮음 | - |
| 이스라엘 | 높음 | 높음 | 높음 | 높음 | 높음 | - |

### 한국에서 서비스 제한의 시사점
- 수치지도 반출 제한 → 구글이 국내 고정밀 지도 데이터 미보유 → 차량·도보·자전거·택시 서비스 전면 미제공
- 대중교통은 제공되나 도로 기반 동선 미반영 (지도 데이터 부재에서 기인)
- 대한민국 IP에서 Google Flights 연결 불가 (타국 IP에서는 가능)
- 공간정보 기본법 개정 시 이들 서비스 확대 가능성 고려 필요
